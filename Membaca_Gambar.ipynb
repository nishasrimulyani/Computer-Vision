```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

#1. Baca gambar
img= cv2.imread('sample_gambar.jpg')

#2. Menampilkan informasi matriks gambar
print("Tipe Data : ", type(img))
print("Ukuran gambar (Tinggi, Lebar, Channer Warna) :", img.shape)

#3. Menampilkan Gambar
cv2_imshow(img)
```

```python
#Konversi dari BGR ke RGB (untuk tampilan yang benar)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Konversi dari BGR ke Grayscale (Hitam Putih)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Menampilkan kedua  gambar berdampingan menggunakan Matplotllib
fig, axs = plt.subplots(1, 2, figsize=(10,5))

axs[0].imshow(img_rgb) #Menampilkan gambar rgb
axs[0].set_title('Gambar RGB') #judul gambar axs[baris ke berapa]
axs[1].imshow(img_gray, cmap='gray') #menampilkan gamber axs[di baris keberapa]
axs[1].set_title('Gambar Grayscale')

plt.show() #menampilkan gambar sampingan
```

```python
#Import library Pillow
from PIL import Image
import numpy as np

# Convert the img_rgb (which is a numpy array) to a PIL Image object
img_pil = Image.fromarray(img_rgb)

#Mengatur area yang di crop
# PIL crop expects (left, upper, right, lower)
# Assuming means (x_start, y_start, width, height)
left, upper, width, height = 200, 200, 500, 600
area_crop_pil = (left, upper, left + width, upper + height)

#Perintah crop
cropped_img_pil = img_pil.crop(area_crop_pil)

#Atur resize
ukuran_img = (400, 700) # (width, height) for PIL resize

#Perintah resize
resized_img_pil = img_pil.resize(ukuran_img)

#Menampilkan kedua  gambar berdampingan menggunakan Matplotllib
fig, axs = plt.subplots(1, 3, figsize=(10,5))

axs[0].imshow(img_rgb)
axs[0].set_title('Gambar RGB')

axs[1].imshow(cropped_img_pil)
axs[1].set_title('Gambar Cropped')

axs[2].imshow(resized_img_pil)
axs[2].set_title('Gambar Resized')

plt.show()
```
